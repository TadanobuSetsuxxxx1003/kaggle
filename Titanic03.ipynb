```python
x = train[['Pclass', 'SibSp', 'Parch']]
y = train[['Survived']]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

```python
x_for_submit = test[['Pclass', 'SibSp', 'Parch']]
submit = test[['PassengerId']]

submit['Survived'] = knn.predict(x_for_submit)

submit
```

```python
x = train[['Pclass', 'SibSp', 'Parch', 'Sex']]  # xにPclassとSexカラムを含むデータフレームを代入
y = train[['Survived']]

x = pd.get_dummies(x, columns=['Pclass', 'Sex'], dtype=int) # pd.get_dummiesを実行
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)

knn.score(x_test, y_test)
```

```python
x = train[['Pclass', 'SibSp', 'Parch', 'Sex', 'Fare']]
y = train[['Survived']]

x = pd.get_dummies(x, columns=['Pclass', 'Sex'], dtype=int)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)

knn.score(x_test, y_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)

x_test_scaled = scaler.transform(x_test)

knn = KNeighborsClassifier()
knn.fit(x_train_scaled, y_train)
knn.score(x_test_scaled, y_test)
```

```python
x_for_submit = test[['Pclass', 'SibSp', 'Parch', 'Sex', 'Fare']]
x_for_submit = pd.get_dummies(x_for_submit, columns=['Pclass', 'Sex'], dtype=int)

x_for_submit_scaled = scaler.transform(x_for_submit)

x_for_submit['Fare'] = x_for_submit['Fare'].fillna(x_for_submit['Fare'].mean())

scaler = StandardScaler()

x_for_submit_scaled = scaler.fit_transform(x_for_submit)

submit = test[['PassengerId']]

submit['Survived'] = knn.predict(x_for_submit_scaled)

submit
```
